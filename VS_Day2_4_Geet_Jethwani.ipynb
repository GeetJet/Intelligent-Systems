## **Question b**
---
Perform the detection of faces, eyes and faces on 'artists.jpg' and display the output.

Suggested parameters:
scaleFactor=1.3,
faceNeighbours=5,
eyeNeighbours=7,
smileNeighbours=28

```
# Type your code below
artist_gray = cv2.cvtColor(artists,cv2.COLOR_BGR2GRAY)
face=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_detect= face.detectMultiScale(artist_gray,scaleFactor=1.3 , minNeighbors = 5)
artists_a = artists.copy()
for (x,y,w,h) in face_detect: 
  cv2.rectangle(artists_a,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(artists_a)
smile = cv2.CascadeClassifier('haarcascade_smile.xml')
smile_detect= smile.detectMultiScale(artist_gray,scaleFactor=1.3 , minNeighbors = 28)

for (x,y,w,h) in smile_detect: 
  cv2.rectangle(artists_a,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(artists_a)
eye = cv2.CascadeClassifier('haarcascade_eye.xml')
eye_detect= eye.detectMultiScale(artist_gray,scaleFactor=1.3 , minNeighbors = 7)

for (x,y,w,h) in eye_detect: 
  cv2.rectangle(artists_a,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(artists_a)

```

Eye detection : The same example of cars/vehicles can be applied here as well . Whenever, the camera is able to detect no eyes ( the driver is dozing off) , it can have a mechanism to park the car itself , slow it down or alert the driver.

Smile detection : Now-a-days due to covid, masks have become mandatory in all public places . Lets take the example of stores or shopping complexes for example, where wearing a mask is mandatory , plus , there are cctvs everywhere . So , in case a person in a shop removes a mask the face becomes evident and persons smile can be identified by the camera , in that case authorities can be alerted that a person has removed their mask.
